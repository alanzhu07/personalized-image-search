```python
def search(model, user, keyword, k=10, retrieve_k=1024):
    # fix unknown keyword embedding
    candidates = [model.image_lookup[image] for image, _ in model.keyword_dict[keyword][:retrieve_k]]
    if len(candidates) < retrieve_k:
        to_sample = np.ma.array(np.arange(len(model.all_images)), mask=False)
        to_sample.mask[candidates] = True
        candidates += np.random.choice(to_sample.compressed(), retrieve_k-len(candidates), replace=False).tolist()
    
    user_id = [model.user_lookup[user]] * len(candidates)
    keyword_id = [model.keyword_lookup[keyword]] * len(candidates)
    x = torch.tensor([user_id, candidates, keyword_id]).T.cuda()
    return x
    # breakpoint()
    # scores = self.model(x).detach().flatten()
    # _, image_ids = torch.topk(scores, k)
    # # breakpoint()
    # return [self.all_images[image_id] for image_id in image_ids.cpu().detach().numpy()]
```
